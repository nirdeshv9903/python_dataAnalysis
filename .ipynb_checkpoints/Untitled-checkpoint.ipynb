```python
anime_data = pd.read_csv('Anime_dataset/anime-dataset-2023.csv')
```

```python
fig = plt.figure(figsize=(15,6))
gs = fig.add_gridspec(3, 3)
ax = fig.add_subplot(222) 
ax.set_title("Rating Chart of Anime", fontweight="bold")
ax.set_xlabel("Anime Shows", color="red")
ax.set_ylabel("Category", color="red")
ax = plt.bar(anime_data["Name"][:20], anime_data["Rating"][:20], color = "red")
ax = plt.grid(linewidth = 0.2, color = "black", alpha = 0.3)
ax = plt.xticks(rotation=45, ha='right', fontsize=7)
ax = plt.yticks(rotation=0, ha='right', fontsize=7)
plt.show()
```
